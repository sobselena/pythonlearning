График

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
x = np.linspace(0, 20, 20)
plt.plot(x, np.tan(x), color='lightblue', marker='8', markersize=5)

ax = plt.gca()
ax.set_xlim(0, 10)
ax.set_ylim(-2, 2)

ax.xaxis.set_minor_locator(ticker.MultipleLocator(1)) 
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.5))  

ax.xaxis.set_minor_formatter(ticker.FormatStrFormatter('%g')) 
ax.yaxis.set_minor_formatter(ticker.FormatStrFormatter('%g')) 

plt.xlabel('Ось х') 
plt.ylabel('Ось y') 
plt.grid(True, linestyle='dashed')
plt.show()
```

Круговая диаграмма

```python
amount = [8, 29, 20, 28, 6, 9]
labels = ["Европа", "Азия", "Африка", "Америка", "Австралия", "Антарктида"]
color_ = ["yellow", "lightgreen", "violet", "lightblue", "pink", "purple"]

plt.pie(amount, labels=labels, colors=color_, autopct='%1.f%%')
plt.title("Площади частей света")
plt.show()
```

Столбчатая диаграмма

```python
x = ["Европа", "Азия", "Африка", "Америка", "Австралия", "Антарктида"]
y = [11.5, 43.4, 30.3, 42, 8.7, 14.1]
ax = plt.gca()
ax.set_ylim(0, 50)
plt.bar(x, y, color="lightgreen")
plt.ylabel('Площадь в млн км^2')
plt.grid(True)
plt.show()
```
